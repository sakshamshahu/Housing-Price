```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("D:\DESKTOP\AML_Project\Housing-Price\data\interim\housing_data.csv\housing_data.csv")
df
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['house_size'], bins=30, kde=True)
plt.title('Distribution of House Size')
plt.xlabel('House Size')
plt.ylabel('Frequency')
plt.ticklabel_format(style='plain', axis='both')  # Remove powers
plt.show()

```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='house_size', y='price', data=df)
plt.title('House Size vs. Price')
plt.xlabel('House Size')
plt.ylabel('Price')
plt.ticklabel_format(style='plain', axis='both')  # Remove powers
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(x='state', y='price', data=df)
plt.title('Price Distribution by State')
plt.xlabel('State')
plt.ylabel('Price')
plt.xticks(rotation=45, ha='right')
plt.ticklabel_format(style='plain', axis='y')  # Remove powers
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='acre_lot', y='price', data=df)
plt.title('Lot Size vs. Price')
plt.xlabel('Lot Size (acres)')
plt.ylabel('Price')
plt.ticklabel_format(style='plain', axis='both')  # Remove powers
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.countplot(x='status', data=df)
plt.title('Count of Houses by Status')
plt.xlabel('Status')
plt.ylabel('Count')
plt.show()
```

```python
numerical_columns = df.select_dtypes(include=['float64', 'int64']).columns
correlation_matrix = df[numerical_columns].corr()
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5, annot_kws={"size": 10})
plt.title('Correlation Matrix Heatmap with Values')
plt.show()
```

```python
plt.figure(figsize=(12, 8))
ax = sns.barplot(x='state', y='price', data=df, ci=None)
plt.title('Average Price by State')
plt.xlabel('State')
plt.ylabel('Average Price')
ax.get_yaxis().get_major_formatter().set_scientific(False)
plt.xticks(rotation=45, ha='right')
plt.show()
```
